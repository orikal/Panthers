```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import os

FILE_PATH = "/content/drive/My Drive/students.txt"

if not os.path.exists(FILE_PATH):
    open(FILE_PATH, "w", encoding="utf-8").close()


def read_students():
    data = []
    with open(FILE_PATH, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue

            parts = line.split(",")

            while len(parts) < 6:
                parts.append("")

            data.append({
                "first": parts[0],
                "email": parts[1],
                "courses": parts[2],
                "link": parts[3],
                "favorite": parts[4]
            })
    return data


students = read_students()


def dropdown_options():
    return [(f"{s['first']}", i) for i, s in enumerate(students)]


title = widgets.HTML(
    "<h2 style='color:#2c3e50;'>Student Information Form</h2>"
)

dropdown = widgets.Dropdown(
    options=dropdown_options(),
    description="Student:",
    layout=widgets.Layout(width="450px")
)

first_name = widgets.Text(
    description="Name:",
    disabled=True,
    layout=widgets.Layout(width="450px")
)

email = widgets.Text(
    description="Email:",
    disabled=True,
    layout=widgets.Layout(width="450px")
)

courses = widgets.Text(
    description="Courses:",
    disabled=True,
    layout=widgets.Layout(width="450px")
)

link = widgets.Text(
    description="Link:",
    disabled=True,
    layout=widgets.Layout(width="450px")
)

favorite_input = widgets.Text(
    description="Favorite Show:",
    placeholder="Enter new favorite show",
    layout=widgets.Layout(width="450px")
)

update_btn = widgets.Button(
    description="Update",
    button_style="success",
    layout=widgets.Layout(width="200px")
)

output = widgets.Output()


def display_student(change=None):
    if not students:
        return

    idx = dropdown.value
    s = students[idx]

    first_name.value = s["first"]
    email.value = s["email"]
    courses.value = s["courses"].replace("|", ", ")
    link.value = s["link"]
    favorite_input.value = s["favorite"]


def update_student(_):
    idx = dropdown.value
    s = students[idx]

    new_value = favorite_input.value.strip()

    if not new_value:
        with output:
            clear_output()
            print("Please enter a value.")
        return

    new_line = (
        f"{s['first']},{s['email']},"
        f"{s['courses']},{s['link']},{new_value}\n"
    )

    with open(FILE_PATH, "a", encoding="utf-8") as f:
        f.write(new_line)

    with output:
        clear_output()
        print("Update saved successfully.")


dropdown.observe(display_student, names="value")
update_btn.on_click(update_student)

display_student()

ui = widgets.VBox([
    title,
    dropdown,
    first_name,
    email,
    courses,
    link,
    favorite_input,
    update_btn,
    output
])

display(ui)
```

```python
import json
import os
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output

file_path = "/content/drive/MyDrive/students.json"

if not os.path.exists(file_path):
    raise FileNotFoundError(f"הקובץ לא נמצא: {file_path}")

def normalize_student(student):
    full_name = student.get("name", "").strip()
    parts = full_name.split()

    first_name = parts[0] if len(parts) >= 1 else ""
    last_name = " ".join(parts[1:]) if len(parts) >= 2 else ""

    return {
        "id": student.get("id", ""),
        "name": full_name,
        "first_name": first_name,
        "last_name": last_name,
        "class": student.get("class", ""),
        "email": student.get("email", ""),
        "courses": student.get("courses", []),
        "interesting_link": student.get("interesting_link", ""),
        "favorite_show": student.get("favorite_show", "")
    }

def load_students():
    with open(file_path, "r", encoding="utf-8") as f:
        raw_data = json.load(f)

    return [normalize_student(s) for s in raw_data]

def save_students(data):
    output_data = []
    for s in data:
        output_data.append({
            "id": s.get("id", ""),
            "name": s.get("name", "").strip(),
            "class": s.get("class", ""),
            "email": s.get("email", ""),
            "courses": s.get("courses", []),
            "interesting_link": s.get("interesting_link", ""),
            "favorite_show": s.get("favorite_show", "")
        })

    with open(file_path, "w", encoding="utf-8") as f:
        json.dump(output_data, f, ensure_ascii=False, indent=2)

students_data = load_students()

def build_dropdown_options():
    return [(s["name"], i) for i, s in enumerate(students_data)]

def courses_to_text(courses):
    if isinstance(courses, list):
        return ", ".join(courses)
    return str(courses)

def text_to_courses(text):
    if not text.strip():
        return []
    return [x.strip() for x in text.split(",") if x.strip()]

# CSS
display(HTML("""
<style>
.form-wrap {
    direction: rtl;
    background: #22252b;
    border: 2px solid #cfcfcf;
    border-radius: 10px;
    padding: 20px;
    width: 720px;
    color: white;
    font-family: Arial, sans-serif;
    margin-top: 10px;
}
.form-title {
    background: linear-gradient(90deg, #4f46e5, #9333ea);
    padding: 16px;
    border-radius: 14px;
    text-align: center;
    font-size: 22px;
    font-weight: bold;
    margin-bottom: 18px;
}
.success-box {
    background: #dcfce7;
    color: #166534;
    padding: 10px;
    border-radius: 8px;
    font-weight: bold;
    margin-top: 10px;
}
</style>
"""))

title = widgets.HTML('<div class="form-title">טופס ניהול סטודנטים</div>')

student_dropdown = widgets.Dropdown(
    options=build_dropdown_options(),
    description='סטודנט:',
    layout=widgets.Layout(width='500px'),
    style={'description_width': '120px'}
)

first_name_text = widgets.Text(
    description='שם פרטי:',
    disabled=True,
    layout=widgets.Layout(width='500px'),
    style={'description_width': '120px'}
)

last_name_text = widgets.Text(
    description='שם משפחה:',
    disabled=True,
    layout=widgets.Layout(width='500px'),
    style={'description_width': '120px'}
)

email_text = widgets.Text(
    description='מייל בראודה:',
    disabled=True,
    layout=widgets.Layout(width='500px'),
    style={'description_width': '120px'}
)

class_text = widgets.Text(
    description='כיתה:',
    disabled=True,
    layout=widgets.Layout(width='500px'),
    style={'description_width': '120px'}
)

courses_text = widgets.Textarea(
    description='קורסים:',
    placeholder='כתוב קורסים מופרדים בפסיקים',
    layout=widgets.Layout(width='500px', height='100px'),
    style={'description_width': '120px'}
)

link_text = widgets.Text(
    description='קישור מעניין:',
    placeholder='הכנס קישור',
    layout=widgets.Layout(width='500px'),
    style={'description_width': '120px'}
)

favorite_show_text = widgets.Text(
    description='תוכנית אהובה:',
    placeholder='הכנס תוכנית אהובה',
    layout=widgets.Layout(width='500px'),
    style={'description_width': '120px'}
)

update_button = widgets.Button(
    description='עדכן ושמור',
    button_style='success',
    layout=widgets.Layout(width='200px')
)

output_msg = widgets.Output()

def display_student(index):
    s = students_data[index]
    first_name_text.value = s.get("first_name", "")
    last_name_text.value = s.get("last_name", "")
    email_text.value = s.get("email", "")
    class_text.value = str(s.get("class", ""))
    courses_text.value = courses_to_text(s.get("courses", []))
    link_text.value = s.get("interesting_link", "")
    favorite_show_text.value = s.get("favorite_show", "")

def on_student_change(change):
    if change["name"] == "value" and change["new"] is not None:
        display_student(change["new"])
        with output_msg:
            clear_output()

def on_update_click(b):
    idx = student_dropdown.value
    if idx is None:
        return

    students_data[idx]["courses"] = text_to_courses(courses_text.value)
    students_data[idx]["interesting_link"] = link_text.value.strip()
    students_data[idx]["favorite_show"] = favorite_show_text.value.strip()

    save_students(students_data)

    with output_msg:
        clear_output()
        display(HTML('<div class="success-box">הנתונים נשמרו בהצלחה בקובץ JSON</div>'))

student_dropdown.observe(on_student_change)
update_button.on_click(on_update_click)

if students_data:
    display_student(0)

form = widgets.VBox([
    title,
    student_dropdown,
    first_name_text,
    last_name_text,
    email_text,
    class_text,
    courses_text,
    link_text,
    favorite_show_text,
    update_button,
    output_msg
])

display(HTML('<div class="form-wrap">'))
display(form)
display(HTML('</div>'))
```
